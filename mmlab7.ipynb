```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
x0=18
N=1655
t=np.arange(0,30,0.1)
```

```python
def func1(t):
    g=0.48
    return g
```

```python
def func_1(t):
    v=0.000081
    return v
```

```python
def func__1(x,t):
    dx=(func1(t)+func_1(t)*x)*(N-x)
    return dx
```

```python
x1=odeint(func__1,x0,t)
```

```python
plt.plot(t,x1,lw=2)
plt.title('Случай 1')
plt.grid('True')
```
